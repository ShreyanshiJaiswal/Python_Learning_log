To get specific rows through indexes we can use the iloc function which is acronym for index location. Also we have given that wanted to see rows from 0 to 5 so we mentioned [0:6] because the last is not not counted so it basically means from 0 until 5.

```python
df.iloc[3:6]
```

Similar as above just index is starting from 3 and ending at 5.

```python
df.iloc[3:]
```

Starting from 3 till end

The above code is used to select the desired columns from the selected row indexes.

```python
df.loc[0:10,['order_id','order_date','sales']]
```

**In loc we can pass the index value through which we can get the actual value that is in the csv file. Almost similar as iloc but the biggest differnce is when we write [0:10]- iloc prints from starting till 9 whereas we can see in the output loc prints till 10 starting from 0 index.** Also [0:10]- it is called as index value

```python
df_new = df.loc[0:10,['order_id','order_date','sales']]
df_new
```
