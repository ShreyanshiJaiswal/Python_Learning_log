# **DROP DUPLICATES**

```python
data = {
    "name":['Sumit','Rahul','Ankit','Ankit'],
    "age":[25,30,30,30],
    "salary":[25000, 30000, 35000,35000]
}
df =pd.DataFrame(data)
df
```

If we look closely we can see we have two exact duplicate rows in the table which is 2 and 3. and based on age 1,2,3 are duplicates.

```python
df.drop_duplicates()
```

It is checking on all the rows which are exact duplicates and what it is doing, it is keeping only the first row and deleting the rest. As by default it the behaviour of drop_duplicate function to keep only the first record.

```python
df.drop_duplicates(keep='last')
```

Now we have changed the behaviour by passing the condition to keep the last duplicate here.

```python
df.drop_duplicates(subset='age')
```

It is keeping the unique ones based on the condition of age.

```python
df.drop_duplicates(subset='age',keep='last')
```

This is giving us the freedom to keep the last entry based upon age.

```python
df.drop_duplicates(subset=['name','age'])
```

We are keeping columns based upon 2 parameters which is name and salary.

```python
df_orders = pd.read_csv('orders.csv')
df_returns = pd.read_csv('returns.csv')
```

What we have done here is that we have created two data frames for both the csv files we have attached to work with.

```python
df_returns
```

```python
df_orders
```

Here we need to look at one most important thing which is that 'return_order_id' from the returns csv is same as the 'order_id' from the orders.csv, both are common columns in both the tables.

```python
df = pd.merge(df_orders,df_returns,how='left',left_on='order_id',right_on='return_order_id')
```

here what we have done we have taken the both dataframes which are 'df_orders' & 'df_returns' and then we passed the arguements how which is basically us telling that we are trying to do left join. Here in our case the left table is the orders and right table is returns.
and since one very important thing here is that - **Since both the common columns are not named as same we have to pass the parameters left_on ='order_id' & right_on='return_order_id' because it is like telling that these ywo are common columns for both the tables**.

```python
df
```

from the above code, we can see that since we have done left join we are getting the columns where there is a return_reason for a certain order_id and where there is not we are getting null values called as '**NaN**' in pandas

```python
df = pd.merge(df_orders,df_returns,how='inner',left_on='order_id',right_on='return_order_id')
```

Here we are doing inner join, we will get those rows that match in both tables, so null values are not shown.

```python
df
```

here we are just getting those where there is a reason for return.

```python
df.groupby('return_reason')['sales'].sum()
```

Here what we have done is we have performed aggregate function to learn the amount we have lost on sales because of return_reasons.

```python
df = pd.merge(df_orders,df_returns,how='cross')
```

this has done cross join that is a cartesian product of two tables.

```python
df
```
